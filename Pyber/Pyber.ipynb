# Pyber Ride Sharing

## Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
#matplotlib inline
```

```python
city_path = "raw_data/city_data.csv"
ride_path = "raw_data/ride_data.csv"

city_data = pd.read_csv(city_path)
ride_data = pd.read_csv(ride_path)

# merge csv files
final_data = pd.merge(city_data, ride_data, how="outer", on="city")
final_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
urban = final_data[(final_data.type == "Urban")]
suburban = final_data[(final_data.type == "Suburban")]
rural = final_data[(final_data.type == "Rural")]

urban_rides_per_city = urban.groupby([final_data.city]).count()["ride_id"]
urban_avg_fare = urban.groupby([final_data.city]).mean()["fare"]

suburban_rides_per_city = suburban.groupby([final_data.city]).count()["ride_id"]
suburban_avg_fare = suburban.groupby([final_data.city]).mean()["fare"]

rural_rides_per_city = rural.groupby([final_data.city]).count()["ride_id"]
rural_avg_fare = rural.groupby([final_data.city]).mean()["fare"]

plt.scatter(urban_rides_per_city, urban_avg_fare, label = "Urban", s=urban_rides_per_city*10, marker="o", c=["lightcoral"], edgecolors="black")
plt.scatter(suburban_rides_per_city, suburban_avg_fare, label = "Suburban",s=suburban_rides_per_city*10, marker="o", c=["lightskyblue"], edgecolors="black")
plt.scatter(rural_rides_per_city, rural_avg_fare, label = "Rural", s=rural_rides_per_city*10, marker="o", c=["gold"], edgecolors="black")

plt.xlim(0, 40)
plt.ylim(15, 52)

plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides (per city)")
plt.ylabel("Average Fare ($)")

plt.legend(title="City Types")
plt.annotate("Note: \nCircle size correlates with driver count per city", xy=(30,30),xycoords="data", xytext=(42.5,35))

plt.grid(color="grey")
plt.show()
```

## Total Fares by City Type

```python
total_rides_per_city_fare = final_data.groupby(["type"])[["fare"]].sum()

plt.pie(total_rides_per_city_fare,
        labels=["Rural","Suburban","Urban"],
       colors=["gold","lightskyblue", "lightcoral"],
       autopct= '%.2f%%',
       explode=[0,0,0.1],
       pctdistance = 0.6,
       shadow=True,
       startangle=120)

plt.title("Percentage of Total Fares by City Type")
plt.axis("equal")
plt.show()
```

## Total Rides by City Type

```python
total_rides_by_city_type = final_data.groupby(["type"])[["ride_id"]].count()

plt.pie(total_rides_by_city_type, 
        explode=[0,0,.1], 
        labels=["Rural","Suburban","Urban"], 
        colors=["gold","lightskyblue", "lightcoral"], 
        autopct= '%.2f%%',
        pctdistance=.6, 
        shadow=True, 
        startangle=120)


plt.title("Percentage of Total Rides by City Type")
plt.axis("equal")
plt.show()
```

## Total Drivers by City Type

```python
total_drivers_by_city_type = city_data.groupby(["type"])[["driver_count"]].sum()

plt.pie(total_drivers_by_city_type, 
        explode=[0,0,.1], 
        labels=["Rural","Suburban","Urban"], 
        colors=["gold","lightskyblue", "lightcoral"], 
        autopct= '%.2f%%',
        pctdistance=.6, 
        shadow=True, 
        startangle=140)

plt.title("% of Total Drivers by City Type")
plt.axis("equal")
plt.show()
```
